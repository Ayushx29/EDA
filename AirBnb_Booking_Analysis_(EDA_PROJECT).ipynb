## 3. ***Data Wrangling***

### Data Wrangling Code

````python
Airbnb.head(10)
````

````python
# Write your code to make your dataset analysis ready.
# 1.] What Is The Average Preferred Price By The Customers Acoording To The Location.
avg_price = Airbnb.groupby(['neighbourhood_group','room_type'],as_index = False)['price'].mean().rename(columns = {'neighbourhood_group':'City','price':'Average_Price'})
avg_price =Airbnb.groupby(['neighbourhood_group','room_type'])['price'].mean().unstack()
avg_price
````

````python
# 2.] No. Of Active Hosts Per Location.
no_host = Airbnb.groupby('neighbourhood_group',as_index = False)['host_id'].count().sort_values(['host_id'], ascending = False).rename(columns={'neighbourhood_group':'City'})
no_host
````

````python
Airbnb[Airbnb['minimum_nights']==1].sort_values(['price','minimum_nights'],ascending = [True ,True])
````

````python
Airbnb.loc[Airbnb['price'] < 100 , 'price'].count()
Airbnb_price= Airbnb.copy()
round(Airbnb.loc[Airbnb['price'] < 100,'price'].mean())
Airbnb_price['price'] = np.where((Airbnb_price['price'] <= 50) & (Airbnb_price['minimum_nights']==1) , round(Airbnb.loc[Airbnb['price'] < 100 , 'price'].mean()), Airbnb_price['price'])
Airbnb_price[Airbnb_price['price']==0].count()
````

````python
Airbnb_price[Airbnb_price['price']==0]
````

````python
def calculate(min_night):
  mean_price = round(Airbnb.loc[Airbnb['price'] < 100 , 'price'].mean())
  price = mean_price * min_night

  return price
````

````python
Airbnb_price['price'] = np.where( (Airbnb_price['price']==0),calculate(Airbnb_price['minimum_nights']) , Airbnb_price['price'] )
Airbnb_price.describe()
````

````python
# 3.] What Is The Highest & Lowest Price of Rent For Customer According To Location.
max_price = Airbnb_price.groupby('neighbourhood_group',as_index = False)['price'].max().sort_values(['price'], ascending = False).rename(columns={'neighbourhood_group':'City','price':'max_price'})
max_price
````

````python
min_price = Airbnb_price.groupby('neighbourhood_group',as_index = False)['price'].min().sort_values(['price'], ascending = True).rename(columns={'neighbourhood_group':'City','price':'min_price'})
min_price
````

````python
merge = pd.merge(max_price , min_price , on = 'City')
merge
````

````python
merge_copy = merge.copy()
def log_values(numbers):
  new_max_price = np.log2(numbers)

  return new_max_price

merge_copy['max_price'] = merge_copy.apply(lambda x: log_values(x['max_price']),axis = 1)
merge_copy['min_price'] = merge_copy.apply(lambda x: log_values(x['min_price']),axis = 1)
merge_copy
````

````python
# 4.] Most Demanded Host Of Airbnb In New York City
host_review = Airbnb_price.groupby(['host_id','host_name'], as_index = False)['number_of_reviews'].sum().sort_values(['number_of_reviews'], ascending = False)
host_review
````

````python
host_avai = Airbnb_price.groupby(['host_id','host_name'], as_index = False)['availability_365'].sum().sort_values(['availability_365'], ascending = False)
host_avai
````

````python
popular_host = pd.merge(host_review , host_avai , on = ['host_id','host_name']).sort_values(['availability_365','number_of_reviews'],ascending = [True,False]).rename(columns = {'number_of_reviews':'Number Of Reviews','availability_365':'Availability'})
popular_host
````

````python
popular_host.head(10)
````

````python
# 5.] Total No. Of Nights Spend Per City.
total_nights = Airbnb_price.groupby('neighbourhood_group')['minimum_nights'].sum().reset_index().sort_values('minimum_nights' , ascending = True)
total_nights
````

````python
#6.] Total No. Of Nights Spend Per Room.

total_nights_room = Airbnb_price.groupby('room_type')['minimum_nights'].sum().reset_index().sort_values('minimum_nights', ascending = False)
total_nights_room
````
